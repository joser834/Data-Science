## Image Operations & Transformations

### Image Binarization - Thresholding

```python
def binarization(image, threshold):
    im = cv2.imread(image)
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    
    height = im_gray.shape[0]
    width = im_gray.shape[1]
    channels = im_gray.shape[2]
    
    im_binary = np.where(im_gray<=threshold,0,255)
    
    #original image
    cv2.imwrite("im_gray.jpg", im_gray)
    gray = cv2.imread("im_gray.jpg")
    cv2.imshow("Original Image", gray)
    
    #result image
    cv2.imwrite("im_binary.jpg", im_binary)
    binary = cv2.imread("im_binary.jpg")
    cv2.imshow("Result Binary Image", binary)
    
    #parameters
    cv2.waitKey(16000)
    cv2.destroyAllWindows()
    cv2.waitKey(1)
    sys.exit()
```

## Aritmetic and Logical Operations

Now, I'll define a function for the following aritmetic and logical operations:

- Addition
- Substraction
- AND
- OR
- XOR

These will be the input images

**Note: Both images have the same size for simplicity purposes**

<img src = 'batman.jpg'/>

<img src = 'triangle.jpg'/>

```python
def binary_image(image, threshold):
    im = cv2.imread(image)
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    
    im_binary = np.where(im_gray<=threshold,0,255)
    return(im_binary)
```

```python
def im_operation(image1, image2, operation):
    
    if operation not in ['addition', 'substraction', 'and', 'or', 'xor']:
        print('Please select an operation from the following: addition, substraction, and, or, xor')
        return(0)
    print(operation)
    
    if operation == 'addition':
        out = binary_image(image1, 180) + binary_image(image2, 180)
        #result image
        cv2.imwrite(operation+'.jpg', out)
        op_image = cv2.imread(operation+'.jpg')
        return(op_image)
    
    elif operation == 'substraction':
        out = binary_image(image1, 180) - binary_image(image2, 180)
        #result image
        cv2.imwrite(operation+'.jpg', out)
        op_image = cv2.imread(operation+'.jpg')
        return(op_image)
    
    elif operation == 'and':
        out = np.where( 
            ((binary_image(image1, 180)==0) & (binary_image(image2, 180)==0)),
            0,
            np.where(
                ((binary_image(image1, 180)==0) & (binary_image(image2, 180)==255)) | ((binary_image(image1, 180)==255) & (binary_image(image2, 180)==0)),
                0,
                255
            )
        )
        #result image
        cv2.imwrite(operation+'.jpg', out)
        op_image = cv2.imread(operation+'.jpg')
        return(op_image)
    
    elif operation == 'or':
        out = np.where( 
            ((binary_image(image1, 180)==0) & (binary_image(image2, 180)==0)),
            0,
            255
        )
        #result image
        cv2.imwrite(operation+'.jpg', out)
        op_image = cv2.imread(operation+'.jpg')
        return(op_image)
    
    elif operation == 'xor':
        out = np.where( 
            ((binary_image(image1, 180)==0) & (binary_image(image2, 180)==0)),
            0,
            np.where(
                ((binary_image(image1, 180)==0) & (binary_image(image2, 180)==255)) | ((binary_image(image1, 180)==255) & (binary_image(image2, 180)==0)),
                255,
                0
            )
        )
        #result image
        cv2.imwrite(operation+'.jpg', out)
        op_image = cv2.imread(operation+'.jpg')
        return(op_image)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
def im_3d(image, elev, rotation):
    im = cv2.imread(image)
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    
    fig = plt.figure()
    ax = Axes3D(fig)
    
    x = im_gray.shape[0]
    y = im_gray.shape[1]
    
    X, Y = np.mgrid[range(0, x),range(0, y)]
    
    ax.view_init(elev=elev, azim=rotation)
    
    ax.plot_surface(X, Y, im_gray, cmap = plt.cm.gray)
    plt.title(image)
```

```python
im_3d('batman.jpg', 30, -60)
```

```python
im_3d('batman.jpg', 90, 0)
```

```python
im_3d('batman.jpg', 60, 0)
```

## Image Transformations

```python
def im_transform(image, transformation, **kwargs):
    if transformation not in ['negative', 'linear', 'logarithmic', 'exponential']:
        print('Please select an operation from the following: negative, linear, logarithmic, exponential')
        return(0)
    print(transformation)
    
    im = cv2.imread(image)
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    
    height = im_gray.shape[0]
    width = im_gray.shape[1]
    
    if transformation == 'negative':
        out = 255-im_gray
        #result image
        cv2.imwrite(transformation+'.jpg', out)
        tnsf_image = cv2.imread(transformation+'.jpg')
        return(tnsf_image)
    
    elif transformation == 'linear':
        print('alpha: ' + str(kwargs['alpha']))
        print('beta: ' + str(kwargs['beta']))
        out = kwargs['alpha']*im_gray + kwargs['beta']
        #result image
        cv2.imwrite(transformation+'.jpg', out)
        tnsf_image = cv2.imread(transformation+'.jpg')
        return(tnsf_image)
    
    elif transformation == 'logarithmic':
        print('alpha: ' + str(kwargs['alpha']))
        out = kwargs['alpha']*np.log(np.ones((height, width))+im_gray)
        #result image
        cv2.imwrite(transformation+'.jpg', out)
        tnsf_image = cv2.imread(transformation+'.jpg')
        return(tnsf_image)
    
    elif transformation == 'exponential':
        print('alpha: ' + str(kwargs['alpha']))
        print('gamma: ' + str(kwargs['gamma']))
        out = kwargs['alpha']*(im_gray**kwargs['gamma'])
        #result image
        cv2.imwrite(transformation+'.jpg', out)
        tnsf_image = cv2.imread(transformation+'.jpg')
        return(tnsf_image)
```

```python
plt.imshow(
    im_transform('batman.jpg', 'negative')
)
```

## Image Convolution

```python
def im_convolution(image, kernel):
    
    if kernel.shape[0] != kernel.shape[1]:
        return('Please use squared matrix for the kernel')
    elif (kernel.shape[0]%2) == 0:
        return('Please use an odd dimension for the kernel')
    
    im = cv2.imread(image)
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    
    height = im_gray.shape[0]
    width = im_gray.shape[1]
    
    start = int(np.floor(kernel.shape[0]/2))
    
    new_image = np.array([])
    
    for i in range(start, width-start-1):
        for j in range(start, height-start-1):
            new_matrix = im_gray[i-int(np.floor((kernel.shape[0])/2)):i+int(np.floor((kernel.shape[0])/2)+1),
                                 j-int(np.floor((kernel.shape[0])/2)):j+int(np.floor((kernel.shape[0])/2)+1)]
            value = (np.sum(new_matrix*kernel))/(kernel.shape[0]*kernel.shape[1])
            new_image = np.append(new_image, value)
    new_image = new_image.reshape((height-kernel.shape[0], width-kernel.shape[0]))
    cv2.imwrite('out.jpg', new_image)
    return(new_image)
```

```python
plt.imshow(im_convolution('nature.jpg', np.matrix([[1,2,1],[2,4,2],[1,2,1]]))) #604 x 340
```
