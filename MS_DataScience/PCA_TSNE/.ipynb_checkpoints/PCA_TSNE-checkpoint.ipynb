```python
import numpy as np
import imageio as imio
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

## PCA and T-SNE

The data is broken down into three different arrays:

* features : Image features by Google Inception
* labels : Dog breed
* images_name : Original image name

```python
images_inception_features = './pre-processing/training_inception_features_array.npy'
training_image_labels_array = './pre-processing/training_image_labels_array.npy'
training_image_names_array = './pre-processing/training_image_names_array.npy'
```

```python
TRAIN_DICTIONARY_KEYS = ["images", "label","features"]

training_set_dict = dict()
    
training_set_dict['images_name'] = np.load(training_image_names_array)
training_set_dict['label'] = np.load(training_image_labels_array)
training_set_dict['features'] = np.load(images_inception_features)

label = np.unique(training_set_dict['label'])
label_index = np.arange(1,len(label)+1)
label_dict = {'label_index': label_index, 'label': label}
label_dict = pd.DataFrame(label_dict, index=label)

new_list = [None]*len(training_set_dict['label'])
for i in range(0,len(training_set_dict['label'])):
    temp = label_dict[label_dict.index==training_set_dict['label'][i]]
    new_list[i] = temp['label_index'][0]
    
training_set_dict['label_code'] = new_list
```

### 1. Principal Component Analysis (PCA) using Sklearn

```python
pca = PCA(n_components=2)
pca_results = pca.fit_transform(training_set_dict['features']) #new features matrix
```

```python
print(training_set_dict['features'].shape)
print(pca_results.shape)
```

```python
plt.scatter(
    pca_results[:, 0], 
    pca_results[:, 1]
)
plt.xlabel('Component 1')
plt.ylabel('Component 2')
```

```python
plt.scatter(
    pca_results[:, 0], 
    pca_results[:, 1],
    edgecolor='none', 
    alpha=0.5,
    cmap=plt.cm.get_cmap('Accent', 10)
)
plt.xlabel('Component 1')
plt.ylabel('Component 2')
```

### K-means Clustering for PCA Analysis

We need to get 120 clusters, one for each dog breed

```python
kmeans_pca = KMeans(n_clusters=120)
kmeans_pca_fit = kmeans_pca.fit(pca_results)
```

```python
clusters_pca = kmeans_pca_fit.predict(pca_results)
```

```python
kmeans_pca.cluster_centers_.shape
```

```python
centers_pca = kmeans_pca.cluster_centers_
```

```python
plt.scatter(
    pca_results[:, 0], 
    pca_results[:, 1],
    edgecolor='none', 
    alpha=0.5,
    cmap=plt.cm.get_cmap('Accent', 10)
)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='black', s=200, alpha=0.5);
```

### 2. T-Distributed Stochastic Neighbouring Entities (t-SNE)

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=10, n_iter=300)
tsne_results = tsne.fit_transform(training_set_dict['features'][1:1000])
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(
    tsne_results[:,0], 
    tsne_results[:,1],
    alpha = 0.7
)
```

### K-means Clustering for T-SNE Analysis

Again, we need 120 different clusters because of our size of dog breeds

```python
kmeans_tsne = KMeans(n_clusters=120)
kmeans_tsne_fit = kmeans_tsne.fit(tsne_results)
```

```python
clusters_tsne = kmeans_tsne_fit.predict(tsne_results)
```

```python
kmeans_tsne.cluster_centers_.shape
```

```python
centers_tsne = kmeans_tsne.cluster_centers_
```

```python
plt.scatter(tsne_results[:, 0], tsne_results[:, 1],
            edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Accent', 10))
plt.scatter(centers_tsne[:, 0], centers_tsne[:, 1], c='black', s=200, alpha=0.5);
```
