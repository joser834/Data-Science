```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Initial Data

<img src='Initial Data.png'/>

```python
Q_1 = np.array([[2, -1, 0],
                [-1, 2, -1],
                [0, -1, 2]])
c_1 = np.array([1, 0, 1])
x_1 = np.array([3, 5, 7])
tolerance_1 = 1e-6
N_1 = 30
```

```python
Q_2 = np.array([[2, -1, 0],
                [-1, 2, -1],
                [0, -1, 2]])
c_2 = np.array([1, 0, 1])
x_2 = np.array([-1, 2, -3])
tolerance_2 = 1e-6
N_2 = 30
```

## Gradient Descent with Exact Step Size

<img src='ExactLR.png'/>

```python
def gd_exact_step(Q, c, x, tolerance, N):
    
    #Defining the initial values
    k=1
    error = 100
    
    #Data frame to store each iteration data
    log = pd.DataFrame(columns = ['Iteration', 'Xk', 'Xk+1', 'Pk', 'Step Size', 'Error'])
    
    while (error > tolerance) & (k <= N):
        x_init = x
        gradient = (np.matmul(Q, x) + c)
        step = np.matmul(gradient.T, gradient) / np.matmul(np.matmul(gradient.T, Q), gradient)
        x = x - step*gradient
        error = np.linalg.norm(gradient)
        
        log_info = {
            'Iteration': k,
            'Xk': x_init,
            'Xk+1': x,
            'Pk': gradient,
            'Step Size': step,
            'Error': error
        }
        log = log.append(log_info, ignore_index=True)
        solution = [k, x, gradient, step, error]
        
        k += 1
    return(log, solution)
```

### Using the First Initial Data

```python
exact_log_1, exact_solution_1 = gd_exact_step(Q_1, c_1, x_1, tolerance_1, N_1)
```

```python
plt.plot(exact_log_1['Error'])
plt.title('Convergence with Step Size = '+str(exact_solution_1[3]))
plt.xlabel('Iteration')
plt.ylabel('Gradient Norm')
```

## Gradient Descent with Fixed Step Size

<img src='FixedLR.png'/>

```python
def gd_fixed_step(Q, c, x, step, tolerance, N):
    
    #Defining the initial values
    k=1
    error = 100
    
    #Data frame to store each iteration data
    log = pd.DataFrame(columns = ['Iteration', 'Xk', 'Xk+1', 'Pk', 'Step Size', 'Error'])
    
    while (error > tolerance) & (k <= N):
        x_init = x
        gradient = (np.matmul(Q, x) + c)
        step = step
        x = x - step*gradient
        error = np.linalg.norm(gradient)
        
        log_info = {
            'Iteration': k,
            'Xk': x_init,
            'Xk+1': x,
            'Pk': gradient,
            'Step Size': step,
            'Error': error
        }
        log = log.append(log_info, ignore_index=True)
        solution = [k, x, gradient, step, error]
        
        k += 1
    return(log, solution)
```

### Using the First Initial Data

```python
#Parameters
plt.figure(figsize=(12,12))
sns.set_style("whitegrid")

#Values for step size
lr = [0.001, 0.01, 0.1, 0.5, 1]

i=1
for s in lr:
    fixed_log_1, fixed_solution_1 = gd_fixed_step(Q_1, c_1, x_1, s, tolerance_1, N_1)
    plt.subplot(3,2,i)
    plt.plot(fixed_log_1['Error'])
    plt.title('Convergence for Step Size ='+str(s))
    plt.xlabel('Iteration')
    plt.ylabel('Gradient Norm')
    plt.tight_layout()
    i+=1
```

### Using the Second Inital Data

```python
#Parameters
plt.figure(figsize=(12,12))
sns.set_style("whitegrid")

#Values for step size
lr = [0.001, 0.01, 0.1, 0.5, 1]

i=1
for s in lr:
    fixed_log_2, fixed_solution_2 = gd_fixed_step(Q_2, c_2, x_2, s, tolerance_2, N_2)
    plt.subplot(3,2,i)
    plt.plot(fixed_log_2['Error'])
    plt.title('Convergence for Step Size ='+str(s))
    plt.xlabel('Iteration')
    plt.ylabel('Gradient Norm')
    plt.tight_layout()
    i+=1
```

## Gradient Descent with Variable Step Size

<img src='VariableLR.png'/>

```python
def gd_variable_step(Q, c, x, tolerance, N):
    
    #Defining the initial values
    k=1
    error = 100
    
    #Data frame to store each iteration data
    log = pd.DataFrame(columns = ['Iteration', 'Xk', 'Xk+1', 'Pk', 'Step Size', 'Error'])
    
    while (error > tolerance) & (k <= N):
        x_init = x
        gradient = (np.matmul(Q, x) + c)
        step = 1/k
        x = x - step*gradient
        error = np.linalg.norm(gradient)
        
        log_info = {
            'Iteration': k,
            'Xk': x_init,
            'Xk+1': x,
            'Pk': gradient,
            'Step Size': step,
            'Error': error
        }
        log = log.append(log_info, ignore_index=True)
        solution = [k, x, gradient, step, error]
        
        k += 1
    return(log, solution)
```

### Using the First Initial Data

```python
variable_log_1, variable_solution_1 = gd_variable_step(Q_1, c_1, x_1, tolerance_1, N_1)
```

```python
plt.plot(variable_log_1['Error'])
plt.title('Convergence')
plt.xlabel('Iteration')
plt.ylabel('Gradient Norm')
```
