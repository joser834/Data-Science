# House Price Prediction

## Description

This project consists in creating different single predictive linear regression models in order to find a house sale price using:

**y = f(x) =  mx +b**

**Where:**
- y = dependent variable
- x = independent variable
- m = slope
- b = intercept

**"x"** and **"y"** will be part of the house dataset, and **"m"** and **"b"** will help us find the predicted value of the house. This means that **"m"** and **"b"** are the parameters of the model.

For the model to be able to predict house prices, these values and parameters are needed:
- y = Sale price of a house.
- x = A characteristic of some house.
- yhat = f(x) = estimated sale price of the house.
- m = slope needed to estimate sale price.
- b = intercept needed to estimate sale price.

The project will gather three different models as described below:
1. (A) Sale price prediction using y = f(x) = mx + b
2. (B) Base model of scikit-learn for linear regression
3. (C) The average of the two models described in 1 and 2

### Model (A): Sale price prediction using y = f(x) = mx + b

For this first approach in understanding house prices, the function we're going to use is **y = f(x) = mx + b**.

The parameters of the model will be:
- **epochs:** number of iterations for the model to run
- **print_error_each:** number of times for the model to print an error
- **lr:** (Learning Rate). Will update the parameters each iteration
- **x:** independent variable (One version of the model with Overall Quality and the other one with 1st Floor Square Feet)
- **y:** dependent variable (sale price)

```python
def model_a_price_prediction(epochs, print_error_each, lr, x, y):
    mat = np.array([x,np.ones(len(x))]).transpose()
    m = 1
    b = 0
    vec = np.array([m,b])
    error = np.array([])
    results = {}
    i=0
    for i in range(0,epochs):
        yhat = np.matmul(mat, vec)
        error = np.append(error,np.array([(1/2)*np.average(((y-yhat)**2))]))
        #if(np.mod(i,print_error_each)==0):
        #    print("error",i,": ",error[i], sep="")
        gradient = np.matmul((yhat-y).transpose(),mat)/len(mat)
        m = m - (lr*gradient[0])
        b = b - (lr*gradient[1])
        vec = np.array([m,b])
        results[i] = np.array(vec)
    return(results,error)
```

#### Error from Model A

The following function graphs the error calculated (y axis) in each iteration (x axis) by the formula:

$$error = {1 \over 2} * {\sum_{i=1}^{n} ((y_i-yhat_i)^2) \over n} $$


```python
def model_a_plot_error(dat, n):
    x = np.arange(0,len(dat[1]),n)
    y = dat[1][0::n]
    plt.plot(x,y, 'ro-')
```

### Model (B): Sale price prediction using Sklearn Linear Regression

This second approach to understanding the house prices will use the library **LinearRegression** from **Sklearn**.

The function will use the following parameters:
- **X:** independent variable (One version of the model with Overall Quality and the other one with 1st Floor Square Feet)
- **y:** dependent variable (sale price)

All from training data.

This function called model_b_price_prediction will return the slope (m), the intercept with the Y-axis (b) and the total error using the formula from the previous model.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def model_b_price_prediction(x,y):
    modelB = LinearRegression()
    modelB.fit(X=x.reshape(-1,1), y=y)
    y_hat = modelB.predict(x.reshape(-1, 1))
    error = (1/2*np.mean((y_hat-y)**2))
    return(modelB.coef_, modelB.intercept_, error)
```

### Model (C): Sale price prediction using an average from Iterative Linear Regression (Model A) and Sklearn Linear Regression (Model B)

The final approach to understanding the house prices will be averaging both the iterative and sklearn models.

The function will use the following parameters:
- **iterative_model:** the iterative model with parameters of choice
- **sklearn_model:** the sklearn model with parameters of choice

This function called model_c_price_prediction will return the slope (m), the intercept with the Y-axis (b) and the total error.

```python
def model_c_price_prediction(iterative_model, sklearn_model,x,y,flag):
    m1 = iterative_model[0][len(iterative_model[0])-1][0]
    b1 = iterative_model[0][len(iterative_model[0])-1][1]
    m2 = sklearn_model[0][0]
    b2 = sklearn_model[1]
    y_hat_a = (m1*x) + b1
    error_a = (1/2)*(np.mean((y_hat_a-y)**2))
    y_hat_b = (m2*x) + b2
    error_b = (1/2)*(np.mean((y_hat_b-y)**2))
    y_hat_c = (y_hat_a+y_hat_b)/2
    error_c = (1/2)*(np.mean((y_hat_c-y)**2))
    if flag==1:
        return(y_hat_a, y_hat_b, y_hat_c)
    elif flag==2:
        return(error_a, error_b, error_c)
    else:
        return("Error. Please select a value of 1 or 2 for flag parameter")
```

## Final results and conclusion

As a result from the three previous models, the following funtion (plot_errors) will show the error comparisson between the models. This function will have the parameters:
- **paramA:** vector of errors coming to be compared.
- **paramB:** vector of errors coming to be compared.

```python
oqA = model_a_price_prediction(epochs=1000, 
                               print_error_each=10,
                               lr=0.0497, 
                               x=test['OverallQual'],
                               y=test['SalePrice'])
oqB = model_b_price_prediction(x=test['OverallQual'],
                               y=test['SalePrice'])
oqC = model_c_price_prediction(oqA, oqB, test['OverallQual'], test['SalePrice'], flag=2)
ffA = model_a_price_prediction(epochs=1000, 
                               print_error_each=10,
                               lr=0.00000125, 
                               x=test['1stFlrSF'],
                               y=test['SalePrice'])
ffB = model_b_price_prediction(x=test['1stFlrSF'],
                               y=test['SalePrice'])
ffC = model_c_price_prediction(ffA, ffB, test['1stFlrSF'], test['SalePrice'], flag=2)
mA = np.array([oqA[1][len(oqA[1])-1], ffA[1][len(ffA[1])-1]])
mB = np.array([oqB[2], ffB[2]])
mC = np.array([oqC[2], ffC[2]])
print(mA, mB, mC)
def plot_errors(paramA, paramB, paramC, x, y):
    plt.bar([0,2],paramA,0.35,label='Model A: Iterative')
    plt.bar([0.35,2.35],paramB,0.35,label='Model B: Sklearn')
    plt.bar([0.70,2.70],paramC,0.35,label='Model C: Average Model')
    plt.xticks([(0.35), (2.35)],['Overall Quality', '1stFloorSF'])
    plt.legend()
```

```python
plot_errors(mA,mB,mC,1,1)
```

After checking the errors from the three models, we can see, as shown in the previous graph, that the lowest error comes from the **Sklearn model**. Both variables used in this project have lower error with this model and higher error with the iterative model (A). The green line (model C) is near the middle between both models as it's an average of both of them.
